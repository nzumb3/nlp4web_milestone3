```python
def write_file(mid, title, plot, genres):
    PATH = "/home/bene/Documents/Git_Repositories/nlp4web_milestone3/data/java_ready/"
    if not os.path.exists(PATH + mid + ".txt"):
        with open(PATH + mid + ".txt", "w+") as f:
            f.write(title + "\n")
            f.write("###BREAK###" + "\n")
            f.write(plot + "\n")
            f.write("###BREAK###" + "\n")
            for g in genres:
                if g != "None":
                    f.write(g + ", ")
```

```python
towrite = data_clean[["tconst", "primaryTitle", "Plot", "Genre_1", "Genre_2", "Genre_3"]].get_values()
for row in towrite:
    write_file(row[0], row[1], row[2], row[3:])
```
